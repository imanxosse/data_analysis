## The .groupby() method

<br>

```python
fortune = pd.read_csv('../Datasets/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
len(sectors)
```

```python
fortune['Sector'].nunique()
```

#### Observation
- `len(sectors)` returns the number of grouping which is exactly the output of the fortune['Sector'].nunique()

```python
sectors.size()
```

```python
fortune['Sector'].value_counts()
```

```python
sectors.first()
```

#### Observation
- `sectors.first()` extracts the very first row from every single one of those groups.

```python
sectors.last()
```

#### Observation
- sectors.last() extracts the last row from every single one of those groups.

```python
sectors.groups
```

#### Observation
- `sectors.group` returns a key,value dictionary in which the *key* indicates the group and *value* indicates reference ids of the subset items in each group

```python
# .loc[24] => 24 is the 'Rank' which is chosen as index, that's why we use .loc[]

fortune.loc[24]
# fortune.loc[45]
```

```python
# .iloc[] => 24 is the 'index' which is automatically generated by pandas, 
#            that's why we use .iloc[]
 

fortune.iloc[24]
```

<a id='The_.agg()_method'></a>

## The .agg() method

<br>

```python
fortune = pd.read_csv('../Datasets/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
sectors.agg({
    'Revenue': 'sum',
    'Profits': 'sum',
    'Employees': 'mean'})
```

#### Observation
- Here we applied sum() and mean() functions on our columns

<br>
<br>


```python
sectors.agg(["size", 'sum', 'mean'])
```

#### Observation
- Here, we want to apply size(), sum() and mean() to each column

<br>
<br>

```python
sectors.agg({
    'Revenue': ['sum', 'median'],
    'Profits': 'sum',
    'Employees': 'mean'})
```

#### Observation
- We can modify our aggregation

<a id='Iterating_through_groups'></a>

## Iterating through groups

<br>

```python
fortune = pd.read_csv('../Datasets/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
sectors['Profits'].max()
```

<br>
<br>

```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_company = data.nlargest(1, 'Revenue') # extract the top 1
    df = df.append(highest_revenue_company)
```

```python
df
```

#### Observation
- This operation helps us to iterate on every single sector, extract the company row with the highest revenue and place the company row in a dataframe

<br>
<br>

Let's do the same exercise but on city base analysis

```python
cities = fortune.groupby('Location')
df = pd.DataFrame(columns=fortune.columns)

for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue_in_city)

df.head(20)
```
