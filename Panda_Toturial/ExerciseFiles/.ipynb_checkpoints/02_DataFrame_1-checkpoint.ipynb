## Differences between shared methods

```python
rev = pd.read_csv('../Datasets/revenue.csv', 
                  index_col='Date')
rev
```

```python
rev.sum()
# rev.sum(axis='index')
```

#### Observation
- The default behavior of axis is `axis=index` which is the addition of all the values in each column.

```python
rev.head(10)
```

```python
rev.sum(axis='columns')
```

#### Observatoin
- If we want to sum up by moving across the column and not the rows, we have to change the `axis=1` or `axis=columns`. So, 1606 is infact the summation of 985, 122, and 499.

<a id='Drop_rows_with_null_values'></a>

## Drop rows with null values

```python
nba = pd.read_csv('../Datasets/nba.csv')
nba.tail()
```

```python
nba.dropna()
```

#### Observation
- `dropna()` method by default removes rows from the data frame that have any Null values at all. In another word, if we even have one Null value in the dataset, the whole row will be removed.
- The reason for this behavior is the `how` parameter in dropna() method which is assigned to `any`.

<br>

<br>

```python
nba.dropna(how='all')
```

##### Observation
- If all the columns in the row are Null, the remove it.

<br>

<br>

```python
nba.dropna(how='all', inplace=True)
```

```python
nba
```

```python
# nba.dropna(axis=1)

nba.dropna(axis='columns')
```

#### Observation
- With `axis=1` or `axis='columns'`, removes any columns in the data frame which has Null value.

<br>

<br>

```python
nba.dropna(subset=['Salary', 'College'])
```

#### Observation
- `subset` parameter in .dropna() can accept a list of string and in that list we can define the column that we want to check for our Null values

<a id='Fill_in_null_values_with_the_.fillna()_method'></a>

## Fill in null values with the .fillna() method

```python
nba = pd.read_csv('../Datasets/nba.csv')
nba.head()
```

```python
nba.fillna(value=0)
```

#### Observation
- The `.fillna()` methods tries to find the Null values in the specified column and the replace the null values with the specified values in the `value` parameter.

```python
nba['Salary'].fillna(value=0, inplace=True)
nba.head()
```

```python
nba['College'].fillna(value='No College')
```

```python
nba['College'].fillna(value = 'No College', inplace=True)
nba.head()
```

<a id='The_.astype()_method'></a>

## The .astype() method

The .astype() methos allows us to cenvert the data type of a series.

```python
nba = pd.read_csv('../Datasets/nba.csv').dropna(how='all')
nba['Salary'].fillna(value = 0, inplace=True)
nba['College'].fillna(value = 'No College', inplace=True)

nba.head()
```

```python
nba.dtypes
```

```python
nba.info()
```

<br>
<br>

Let's convert "Salary" from "float64" to "int"

```python
nba['Salary'] = nba['Salary'].astype('int')
```

#### Observation 
- astype() does not have the inplace() parameter

```python
nba.info()
```

```python
nba.head()
```

```python
nba['Age'] = nba['Age'].astype('int')
nba['Number'] = nba['Number'].astype('int')
```

```python
nba.info()
```

<br>

```python
# nba['Position'].unique()
nba['Position'].nunique()
```

```python
nba['Position'] = nba['Position'].astype('category')
```

#### Observation
- The `category` data types is only iavailable in pandas. It's not defined within the python.

- The category data type can help us severly reduce the memory usage

```python
nba.info()
```

#### Observation
- It is obvious that the memory usage has been decreaased in comparision with the time before converting "Position" to "Category"

```python
nba['Team'].nunique()
```

```python
nba['Team'] = nba['Team'].astype('category')
```

```python
nba.info()
```

<a id='Sort_a_dataframe_with_the_.sort_values()_method_Part_I'></a>

## Sort a dataframe with the .sort_values() method, Part I

```python
nba = pd.read_csv('../Datasets/nba.csv')
nba.head()
```

```python
nba.sort_values(by='Name', ascending=False)
```

#### Observation
- Sort the entire data frame by the column "Name"

```python
nba.sort_values(by='Age', ascending=False).head(10)
```

```python
nba.sort_values(by='Salary', ascending=False, inplace=True)
nba.head()
```

#### Note:
By default NaN values are the the end of the data frame but, by using `na_position='first'`, it shows NaN values at first.

```python
nba.sort_values(by='Salary', 
                ascending=False, 
                na_position='first').head(20)
```

<a id='Sort_a_dataframe_with_the_.sort_values()_method_Part_II'></a>

## Sort a dataframe with the .sort_values() method, Part II

```python
nba = pd.read_csv('../Datasets/nba.csv')
nba.head()
```

```python
nba.sort_values(by=['Team', 'Name']).head(10)
```

```python
nba.sort_values(by=['Team', 'Name'], ascending=False).head(10)
```

Lets use different items for ascending parameter.

```python
nba.sort_values(by=['Team', 'Name'], ascending=[True, False]).head(10)
```

#### Observation
- Sort the data By `Team=ascending` and `Name=descending`

<br>
<br>

```python
nba.sort_values(by=['Team', 'Name'], 
                ascending=[True, False], 
                inplace=True)

nba.head()
```
